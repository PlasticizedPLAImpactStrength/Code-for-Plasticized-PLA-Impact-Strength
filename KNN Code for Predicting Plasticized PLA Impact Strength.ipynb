# **KNN Code for Predicting Plasticized PLA Impact Strength**

```python
## Import related libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import neighbors
from math import sqrt
from sklearn.metrics import mean_squared_error,mean_absolute_error,explained_variance_score, r2_score
```

```python
## Import dataset
df = pd.read_excel('Dataset PLA-Plasticizer Bersih 54 Data.xlsx')
df
```

```python
## Define the features and the output (Impact Strength)
X = df[['PLA Mol Mass','Plasticizer Mol Mass','Plasticizer Density','Plasticizer Content']]
y = df[['Impact Strength']]
```

```python
X
```

```python
y
```

```python
## Split the data set into training set and test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=42)
```

```python
## Data normalization
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler_X = MinMaxScaler()
scaler_y= MinMaxScaler()
# Fit and transform the training set
X_train = scaler_X.fit_transform(X_train)
y_train = scaler_y.fit_transform(y_train)
# Fit and transform the test set
X_test = scaler_X.transform(X_test)
y_test = scaler_y.transform(y_test)
```

# Model Building

```python
# Build and set KNN object and parameter
model = neighbors.KNeighborsRegressor(n_neighbors=1)
# Fit the model into the training set
model.fit(X_train, y_train)
```

```python
## Model prediction using test set
model_predictions = model.predict(X_test)
model_predictions = np.array(model_predictions).reshape(-1,1)
```

```python
# Inverse transform the normalized prediction
unnormalized_model_predictions = scaler_y.inverse_transform(model_predictions)
true_y_test = scaler_y.inverse_transform(y_test)
```

```python
## Model performance evaluation
# R2 score
model_r2 = r2_score(true_y_test,unnormalized_model_predictions)

print(model_r2)
```

```python
# RMSE
model_rmse = np.sqrt(mean_squared_error(y_test,model_predictions))
is_rmse = np.sqrt(mean_squared_error(y_test[:,0],model_predictions[:,0]))

print(model_rmse,is_rmse)
```

```python
# RRMSE
def rrmse(true_y_test,unnormalized_model_predictions):
  return np.sqrt(mean_squared_error(true_y_test,unnormalized_model_predictions))/np.average(true_y_test)*100

model_rrmse = rrmse(true_y_test,unnormalized_model_predictions)
is_rrmse = rrmse(true_y_test[:,0],unnormalized_model_predictions[:,0])

print(model_rrmse,is_rrmse)
```
